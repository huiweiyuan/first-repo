```python
#numpy作业
import numpy as np
persontype = np.dtype({
    'names':['name', 'chinese', 'english', 'math'],
    'formats':['S32','i', 'i', 'i']})
peoples = np.array([("ZhangFei",66,65,30),("GuanYu",95,85,98),
       ("ZhaoYun",93,92,96),("HuangZhong",90,88,77),("DianWei",80,90,90)], dtype=persontype)
chineses = peoples[:]['chinese']
englishs = peoples[:]['english']
maths = peoples[:]['math']
# 1. 第一个列表：存储「科目名称」和「对应成绩数组」（一一对应）
subjects = [
    ("语文", chineses),
    ("英语", englishs),
    ("数学", maths)
]

# 2. 第二个列表：存储「统计量名称」和「对应numpy计算函数」（一一对应）
statistics = [
    ("均值", np.mean),
    ("最小值", np.min),
    ("最大值", np.max),
    ("方差", np.var),
    ("标准差", np.std)
]
# -------------------------- 两层for循环：遍历所有组合 --------------------------

# 外层循环：遍历每个科目（拿到科目名+该科成绩）
for subj_name, subj_scores in subjects:
    print(f"===== {subj_name} 相关统计 =====")  # 可选：分组打印，更易读
    # 内层循环：遍历每个统计量（拿到统计量名+计算函数）
    for stat_name, stat_func in statistics:
        # 计算结果：用当前统计函数处理当前科目的成绩
        result = stat_func(subj_scores)
        # 打印结果（格式：统计量名 + 数值，保留2位小数更整洁）
        print(f"{stat_name}：{result:.2f}")
    print()  # 可选：科目之间空一行，区分更清晰

total_scores = chineses + englishs + maths
# 获取总成绩降序排序的索引（argsort默认升序，[::-1]反转为降序）
# argsort返回的是「排序后元素在原数组中的索引位置」
sorted_indices = np.argsort(total_scores)[::-1]
# 根据排序索引重排原始数据，得到按总成绩降序的结果
sorted_peoples = peoples[sorted_indices]
# -------------------------- 格式化输出名次 --------------------------
print("=" * 50)
print(f"{'名次':<6}{'姓名':<10}{'语文':<6}{'英语':<6}{'数学':<6}{'总成绩':<6}")
print("-" * 50)

# 遍历排序后的数据，enumerate从1开始计数（表示名次）
for rank, person in enumerate(sorted_peoples, start=1):
    # 注意：name是S32格式（bytes类型），需要解码为字符串
    name = person['name'].decode('utf-8')
    chinese = person['chinese']
    english = person['english']
    math = person['math']
    total = chinese + english + math  # 也可以用 total_scores[sorted_indices[rank-1]]

    # 格式化输出（左对齐，保证排版整齐）
    print(f"{rank:<6}{name:<10}{chinese:<6}{english:<6}{math:<6}{total:<6}")
print("=" * 50)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
#准备数据集
iris = load_iris()
#获取特征集和分类标识
features = iris.data
labels = iris.target
#随机抽取33%的数据作为测试集（预测），其余为训练集
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.33, random_state=0)
#创建CART分类树
clf = DecisionTreeClassifier(criterion='gini')
#拟合构造CART分类树，将训练集的特征值和分类标识作为参数拟合
clf = clf.fit(train_features, train_labels)
#用CART分类树做预测
test_predict = clf.predict(test_features)
#预测结果与测试集结果作比对
score = accuracy_score(test_labels, test_predict)
print("CART分类树准确率 %.4lf" % score)
```

```python
###预测的关键流程
##模块1：数据探索
import pandas as pd
#数据加载
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
#数据探索
print(train_data. info()) #了解基本情况：行数、列数、每列的数据类型、数据完整度
print('-' *30)
print(train_data. describe()) #了解统计情况：总数、平均值、标准差、最小值、最大值等
print('-' *30)
print(train_data. describe(include=['O'])) #查看字符串类型（非数字）的整体情况
print('-' *30)
print(train_data. head())
print('-' *30)
print(train_data. tail())
print('-' *30)

##模块2：数据清洗
#使用平均年龄填充年龄中的nan值
train_data['Age'].fillna(train_data['Age'].mean(), inplace=True)
test_data['Age'].fillna(test_data['Age'].mean(), inplace=True)
#使用票价的均值填充票价中的nan值
train_data['Fare'].fillna(train_data['Fare'].mean(), inplace=True)
test_data['Fare'].fillna(test_data['Fare'].mean(), inplace=True)
print(train_data['Embarked'].value_counts()) #Embarked为登陆港口，有少量缺失值，可以补齐，先观察取值
#一共3个登陆口，其中S港口人数最多，占到了72%，因此将其余确实的Embarked数值均设为S
train_data['Embarked'].fillna('S', inplace=True)
test_data['Embarked'].fillna('S', inplace=True)

##模块3：特征选择
#通过观察，放弃乘客编号、姓名、cabin（缺失字段多）、船票号作为分类；选择船票等级、性别、年龄、亲戚数量及船票价格交给分类器
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
train_features = train_data[features]
train_lables = train_data['Survived']
test_features = test_data[features]
#特征值里有字符串需要转成数值，使用sklearn特征选择中的DictVectorizer类，处理符号化对象，把符号转化成数字0/1
from sklearn.feature_extraction import DictVectorizer
dvec = DictVectorizer(sparse=False)
train_features = dvec.fit_transform(train_features. to_dict(orient='record'))
print(dvec.feature_names_) #查看dvec在转化后的特征属性是什么样的

##模块4：决策树模型
from sklearn.tree import DecisionTreeClassifier
#构造ID3决策树
clf = DecisionTreeClassifier(criterion='gini')
#决策树训练
clf = clf.fit(train_features, train_lables)

##模块5：模型预测&评估
test_features = dvec.transform(test_features). to_dict(orient='record')
#决策树预测
pred_labels = clf.predict(test_features)
#得到决策树准确率
acc_decision_tree = round(clf.score(train_features, test_labels), 6)
print(u'score准确率为 %。4lf' % acc_decision_tree)
```
